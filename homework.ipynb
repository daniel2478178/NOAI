```python
def factorial(n):
    if n == 1:
        return 1
    return n*factorial(n-1)
inp = int(input().strip())
factorial(inp)
```

```python
class InputString:
    def __init__(self):
        self.s = ""
    def getString(self):
        self.s = input()
    def printString(self):
        print(self.s.upper())
        
a = InputString()
a.getString()
a.printString()

```
